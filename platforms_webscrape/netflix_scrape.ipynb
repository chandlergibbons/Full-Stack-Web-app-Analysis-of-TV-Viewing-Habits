```python
from bs4 import BeautifulSoup as bs
import pandas as pd
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import requests
import time
import re
import time
import random
```

Netflix Scrape

```python
show_title = []
show_year = []
imdb_rating = []
real_rating =[]
prime = []
netflix = []
disney = []
hulu = []
```

```python
url = 'https://reelgood.com/tv/source/netflix?filter-year_end=2021&filter-year_start=2016'
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
browser.visit(url)
```

```python
try:
    while True:
        time.sleep(float("{0:.2f}".format(random.uniform(1, 3))))
        browser.click_link_by_partial_text('Load More')
except:
    print("ready to scrape")
```

```python
html = browser.html
soup = bs(html, 'html.parser')

title_col = (soup.find_all('td', class_='css-1u7zfla e126mwsw1'))

e_else_col = (soup.find_all('td', class_='css-1u11l3y'))


```

```python
for i in range(len(e_else_col)):
    if i % 4 == 0:
        show_year.append(str(e_else_col[i]).split("</")[0].split(">")[1])
       
```

```python
len(show_year)
```

```python
for i in range(2, len(e_else_col)):
    if (i + 2) % 4 == 0:
        try:
            imdb_rating.append(str(e_else_col[i]).split("</b")[0].split(">")[5])
        except:
            imdb_rating.append(str(e_else_col[i]).split("</b")[0].split(">")[3])
```

```python
for i in range(3, len(e_else_col)):
        if (i + 1) % 4 == 0:
            try:
                real_rating.append(str(e_else_col[i]).split("</b")[0].split(">")[5])
            except:
                real_rating.append(str(e_else_col[i]).split("</b")[0].split(">")[3])
    
```

```python
for i in range(len(title_col)):
    show_title.append((str(title_col[i])).split("</")[0].split("\">")[2])
```

```python
for i in range(len(show_title)):
    netflix.append(True)
```

```python
for i in range(len(show_title)):
    prime.append(False)
    disney.append(False)
    hulu.append(False)
    
```

```python
netflix_whole = [show_title, show_year, imdb_rating, real_rating, prime, netflix, disney, hulu]
```
